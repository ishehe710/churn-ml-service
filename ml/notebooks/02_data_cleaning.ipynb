<h1>Data Cleaning</h1>

```python
# load raw dataset
import pandas as pd

df = pd.read_csv("../../data/raw/telco_churn.csv")
df.tail()
```

<h2>Cleaning 'Total Charges' Column</h2>

```python
# replacing whitespace
df['TotalCharges'] = df['TotalCharges'].replace(' ', pd.NA)

# chaning to a numeric value
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'])

# confirmation
df['TotalCharges'].isna().sum()
df['TotalCharges'].dtype

# removing rows with Nan for 'Total Charges'
df = df.dropna(subset=['TotalCharges'])

df.tail()
```

<h2>Cleaning 'Churn' Column</h2>

```python
# making the 'Churn' column numeric
df['Churn'] = df['Churn'].map({'Yes': 1, 'No':0})

# check
df['Churn'].value_counts()
df['Churn'].dtype
df.tail()
```
