```python
# checking for missing data

df.isna().sum().sort_values(ascending=False)

keys = df.keys()

for key in keys:
    key_type = df[key].dtype
    num_whitespace_strings = (df[key] == ' ').sum()
    print(f"{key} is of type {key_type} and has {num_whitespace_strings} empty strings.")
```

<p>Notes</p>
<ul>
    <li>All columns have values, non Nones or Nans</li>
    <li>Total Charges as empty strings as values. Pandas does not automatically check for empty strings.</li>
    <li>Total Charges will require cleaning</li>
</ul>
